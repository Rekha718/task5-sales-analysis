Load CSV

```python
import pandas as pd

df = pd.read_csv("sales.csv")
print(df.head())
```

Total Revenue by Region

```python
revenue_by_region = df.groupby("Region")["Total Revenue"].sum()
print(revenue_by_region)
```

Total Profit by Country

```python
profit_by_country = df.groupby("Country")["Total Profit"].sum().sort_values(ascending=False).head(10)
print(profit_by_country)
```

Plotting Examples

```python
import matplotlib.pyplot as plt

# Bar chart: Revenue by Region
revenue_by_region.plot(kind="bar", title="Total Revenue by Region")
plt.ylabel("Revenue")
plt.show()

# Top 10 Profitable Countries
profit_by_country.plot(kind="bar", title="Top 10 Profitable Countries")
plt.ylabel("Profit")
plt.show()
```
